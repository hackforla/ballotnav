```python
path = './CEB_Physical Addresses_08072020.pdf'
doc = fitz.open(path)
```

```python
doc
```

```python
raw_text = ""
for page in doc:
    text = page.getText("text")
    raw_text += text
```

```python
len(raw_text)
```

```python
raw_text[:1000]
```

```python
raw_text.split('\n')[:12]  # the first 10 lines are the headers
```

```python
raw_new = raw_text.split('\n')[10:] #removing the first 10 lines
```

```python
raw_new[364:374]  #Headers on the second page! - the headers have to be removed to make the nth step work
```

```python
raw_new4 = raw_new[:364] + raw_new[374:] #this works!
```

```python
raw_new4[340:380] #this works
```

```python
len(raw_new4)
```

```python
# Removing the asterix symbol, which occurs 7 times throughout the string and throws off the nth step split
# while removes all instances of *

while "*" in raw_new4: raw_new4.remove("*") 
```

```python
len(raw_new4)
```

```python
raw_new4[100:155]
```

```python
# Splitting list by county - every county happens to have 9 items in its contact information. Eureka!
# Finally a beautiful list with the correct structure! iterable!

n=9
[raw_new4[i:i+n] for i in range(0, len(raw_new4), n)]
```

```python
counties = [raw_new4[i:i+n] for i in range(0, len(raw_new4), n)]
```

#### Main function

```python
list_1 = []

for item in counties[:77]:
    dict = {
        'State' :'OK',
        'County' : item[8][2:].strip(),
        'location_type' : 'County Election Board',
        'Secretary' : item[3].strip(),
        'address_1' : item[2].strip(),
        'address_2' : item[4].strip(),
        'city' : '',
        'state' : '',
        'phone' : item[0].strip(),
        'latitude' :'',
        'longitude' : '',
        'has_dropoff_location' : 'True',
        'has_phone_number' : 'True',
        'county_website_url' : '',
        'email' : '',
        'fax' : item[1].strip(),
        'social' : '',
        'inactive' : '',
        'hours' : item[7] + item[6] + item[5],
        'rules' : 'Applications for absentee ballots must be made in writing or using the Oklahoma State Election Board\'s Online Absentee Voting Application. Absentee ballot application forms are available from all county election boards and from the State Election Board.',
        'notes' : ''
    }
    list_1.append(dict)
    
#print(list_1) - this works
pd.DataFrame(list_1)
```

```python
counties_df = pd.DataFrame(list_1)
counties_df
```

```python
counties_df.to_csv('Oklahoma County Location.csv', index=False)
```
